- **Как прочитать данные**
  
  Воспользуйтесь функцией `read_csv` из пандаса, чтобы считать данные по ссылке к ним
  
  Например,
  
  ```python
  df = pd.read_csv('[<https://my.site/data.csv>](<https://my.site/data.csv>)')
  ```
  
  Считает данные, хранящиеся по ссылке `https://my.site/data.csv`
  
- **Как делать группировку?**
  
  Для этого существует метод `groupby` - его вызов группирует строки по значениям в какой-то колонке (или комбинации значений из нескольких колонок). Например,
  
  ```python
  df.groupby('pickup_month')
  ```
  
  Сгруппирует данные датафрэйма `df` по колонке `pickup_month` - то есть строки с месяцем январь в колонке `pickup_month` будут в отдельной колонке от строк с месяцем май. Сама по себе группировка не вычисляет что-то по группам
  
- **Что делать после группировки**
  
  После группировки нужна агрегация методом `agg` или применение `apply` (об этом будет позже). Чтобы посчитать общее число поездок в каждой из групп, используйте метод `agg` с функцией `sum`. А затем возьмите район с максимальным числом поездок
  
- **Как сделать агрегацию?**
  
  Метод `agg` принимает словарь, где в качестве ключей указаны названия колонок, а значениями являются функции, которые нужно применить к колонке-ключу в каждой группе. К примеру, этот код
  
  ```python
  df.groupby('pickup_month') \\
  .agg({'temp': 'sum'})
  ```
  
  Сгруппирует строки по значению в колонке `pickup_month` (месяц поездки) и посчитает в каждой группе сумму по колонке `temp`
  
- **Создание функций**
  
  Чтобы сделать функцию напишите её `заголовок` и `тело`
  
  ```python
  заголовок:
      тело
  ```
  
  Пример заголовка:
  
  ```python
  def func_name(parameter):
  ```
  
  Подробнее о том, что здесь написано в конспекте
  
  Пример тела функции:
  
  ```python
  multiplied = parameter * 10
  return multiplied
  ```
  
  Тело функции - это любой код с переменными, которые заданы как параметры в заголовке функции (parameter в примере в круглых скобках).
  
  `return` возвращает из функции какое-то значение так, что вызов функции можно записать в переменную (ниже об этом)
  
  Работа с функциями состоит из 2-ух частей - определение функции и её вызовы. Определение - это создание функции, а вызов - её использование
  
  Создание функции:
  
  ```python
  # Function definition
  def func_name(parameter):
      multiplied = parameter * 10
      return multiplied
  ```
  
  Вызов функции:
  
  ```python
  beautiful_numbers = 10
  
  # Function call
  # beautiful_numbers == parameter
  result = func_name(beautiful_numbers) # multiplied from function will be returned
  ```
  
  В данном примере `beautiful_numbers` является параметром для вызова функции
  
  Результатом вызова функции является переменная `result`
  
  ```python
  result = 100
  ```
  
- **Моя функция ничего не выдаёт!**
  
  А вы добавили в функцию `return`? Добавьте его в конце и напишите там то, что вы хотите вернуть из функции
  
- **Посмотреть разбор**
  
  Посмотреть разбор задания можно по ссылке [](https://youtu.be/2ll9K17G-sk)[3.7 Функция, агрегирующая данные из файла - YouTube](https://youtu.be/2ll9K17G-sk)

````python
def read_n_agg(pan):
    df = pd.read_csv(pan, sep=';')
    res = df.groupby('company')\
            .agg({'income': 'mean'})
    return res                                        

read_n_agg("https://stepik.org/media/attachments/lesson/359209/companies.csv")
````
